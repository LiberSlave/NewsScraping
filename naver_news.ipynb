```python
from newspaper import Article
from konlpy.tag import Okt
from collections import Counter
import time

# 기사 본문 추출 함수
def extract_article_text(url):
    try:
        article = Article(url, language='ko')
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        print(f"Error extracting article from {url}: {e}")
        return ""

# 뉴스 기사 URL 리스트 (여기에 100개 이상의 URL을 추가하세요)
urls = [
    "https://www.g-enews.com/ko-kr/news/article/news_all/2025022014304610224a01bf698f_1/article.html",
    "https://www.etoday.co.kr/news/view/2447223",
    "https://www.thescoop.co.kr/news/articleView.html?idxno=305025"
]

# KoNLPy Okt 객체 생성 (다른 분석기를 사용해도 무방합니다)
okt = Okt()

# 각 기사에서 단어가 등장한 문서 수를 기록할 Counter (문서 빈도)
doc_frequency = Counter()

for idx, url in enumerate(urls):
    print(f"Processing article {idx+1}/{len(urls)}: {url}")
    text = extract_article_text(url)
    if not text:
        continue
    # 형태소 분석기를 사용해 명사 추출 (필요에 따라 다른 POS나 전처리를 할 수 있습니다)
    words = okt.nouns(text)
    # 한 기사 내에서 중복된 단어는 한 번만 세도록 set으로 변환
    unique_words = set(words)
    # 각 단어가 해당 기사에 등장했으므로 document frequency를 업데이트
    doc_frequency.update(unique_words)
    # 너무 빠른 요청으로 서버에 부담을 주지 않도록 잠시 대기 (선택 사항)
    time.sleep(1)

# 예시: 특정 단어의 문서 빈도를 확인
print("\n문서 빈도 예시:")
print(f"'증시'는 {doc_frequency.get('증시', 0)}개의 기사에서 등장")
print(f"'뉴스'는 {doc_frequency.get('뉴스', 0)}개의 기사에서 등장")

# 상위 10개 단어 출력 (문서 빈도가 높은 순)
print("\n상위 10개 단어 (문서 빈도 기준):")
for word, count in doc_frequency.most_common(40):
    print(f"{word}: {count}")
```
